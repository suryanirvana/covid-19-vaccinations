```python
import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup as BS
```

```python
url = "https://www.worldometers.info/world-population/population-by-country/"
request = requests.get(url)
```

```python
countries = BS(request.content).find_all("table")[0]
```

```python
df = pd.read_html(str(countries))[0]
df
```

```python
df = df.drop(columns=["#", "Yearly Change", "Net Change", "Density (P/Km²)", "Land Area (Km²)", "Migrants (net)", "Fert. Rate", "Med. Age", "Urban Pop %", "World Share"], axis=1)
df.head()
```

```python
df.columns = ['Country', 'Population']
df.head()
```

```python
df.to_csv("Country Population.csv", index=None, header=True)
```
